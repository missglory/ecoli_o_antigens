```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

def draw_hist(csv):
    df = pd.read_csv(csv, index_col=0)
    dfn = df.to_numpy()
    dfn = dfn[np.tril_indices(dfn.shape[0], k=1)]
    plt.hist(dfn, bins=100)
    plt.show()

def draw_linkage(csv):
    df = pd.read_csv(csv, index_col=0)
    df = df.to_numpy()
    linkage_matrix = linkage(df, "single") 
    fig = plt.figure(figsize=(20, 20))
    ax = fig.add_subplot(1, 1, 1)
    dendrogram(linkage_matrix)
    ax.tick_params(axis='x', which='major', labelsize=7)
    ax.tick_params(axis='y', which='major', labelsize=10)
    fig.show()
```

# Hists

```python
draw_hist("gedlib_edit_dists.csv")
```

```python
draw_hist("gedlib_lower.csv")
```

```python
draw_hist("gedlib_upper.csv")
```

# Linkage

```python
draw_linkage("gedlib_edit_dists.csv")
```

```python
draw_linkage("gedlib_lower.csv")
```

```python
draw_linkage("gedlib_upper.csv")
```
